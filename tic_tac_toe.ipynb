```python
def Get_Player_Inputs():
    p1 = input(prompt = 'Player 1: Enter X or O: ')
    p1 = p1.upper()
    if p1 == 'X':
        p2 = 'O'
    elif p1 == 'O':
        p2 = 'X'
    else:
        print('Player 1 entered an incorrect choice. Only accepted choices are X or O. It is tic-tac-toe for God\'s sake')
        p1 = 'NULL'
        p2 = 'NULL'
    print ('Player 1 : {} \nPlayer 2 : {}'.format(p1,p2))
    return((p1,p2))
```

```python
def Print_Grid(A):
    print(' {} | {} | {} '.format(A[2][0],A[2][1],A[2][2]))
    print('----o----o---')
    print(' {} | {} | {} '.format(A[1][0],A[1][1],A[1][2]))
    print('----o----o---')
    print(' {} | {} | {} '.format(A[0][0],A[0][1],A[0][2]))
```

```python
def check_if_game_is_over(A,playerInputs):
    #rows
    if (A[0][0] == A[0][1] and A[0][0] == A[0][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    elif (A[1][0] == A[1][1] and A[1][0] == A[1][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    elif (A[2][0] == A[2][1] and A[2][0] == A[2][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    #columns
    elif (A[0][0] == A[1][0] and A[0][0] == A[2][0]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    elif (A[0][1] == A[1][1] and A[0][1] == A[2][1]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    elif (A[0][2] == A[1][2] and A[0][2] == A[2][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    #diagnonal
    elif (A[0][0] == A[1][1] and A[0][0] == A[2][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    elif (A[2][0] == A[1][1] and A[2][0] == A[0][2]):
        if A[0][0] == playerInputs[0]:
            print('Player 1 WON')
        else:
            print('Player 2 WON')
        return(True)
    else:
        return(False)
```

### Main Script - Tic Tac Toe

```python
playerInputs = Get_Player_Inputs()
Print_Grid()
game_over_flag = False
A = [[11,12,13],[21, 22, 23],[31,32,33]]
num_squares_filled = 0
which_player = 0
while game_over_flag == False and num_squares_filled < 9:
    if which_player == 0:
        pos = input('Player 1 input position: ')
    else:
        pos = input('Player 2 input position: ')
        
    row_num = int(pos[0]) - 1
    col_num = int(pos[1]) - 1
    
    A[row_num][col_num] = playerInputs[which_player]
    num_squares_filled = num_squares_filled + 1
    Print_Grid(A)
    game_over_flag = check_if_game_is_over(A,playerInputs)
    if which_player == 0:
        which_player = 1
    else:
        which_player = 0
if num_squares_filled == 9 and game_over_flag == False:
    print('DRAW GAME')
```
